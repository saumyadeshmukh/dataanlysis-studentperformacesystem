```python
# data cleaning

#1. replacing missing value with median for numerical attributes

median=df['family size'].median()
df['family size'].fillna(median,inplace=True)
```

```python
median=df['difficulty level of subjects(1-very low to 5-very high)'].median()
```

```python
df['difficulty level of subjects(1-very low to 5-very high)'].fillna(median,inplace=True)
```

```python
#2. using one hot encoding for categorical attributes

df['father\'s job'].value_counts()
```

```python
df1=pd.get_dummies(df['father\'s job'],drop_first=True)

df=pd.concat([df1,df],axis=1)
```

```python
df=df.drop('father\'s job',axis=1)
```

```python
df=df.rename(columns={'health care related':'f_health care related','other':'f_other'})
```

```python
df2=pd.get_dummies(df['gender'],drop_first=True)

df=pd.concat([df2,df],axis=1)

df=df.rename(columns={'Male':'f_gender'})

df=df.drop('gender',axis=1)
```

```python
df['mother\'s education'].value_counts()

df3=pd.get_dummies(df['mother\'s education'],drop_first=True)

df=pd.concat([df3,df],axis=1)

df=df.rename(columns={'higher education':'m_higher education','none':'m_none','primary education (4th grade)':'m_primary education (4th grade)','secondary education':'m_secondary education'})

df=df.drop('mother\'s education',axis=1)
```

```python
df['father\'s education'].value_counts()
```

```python
df4=pd.get_dummies(df['father\'s education'],drop_first=True)

df=pd.concat([df4,df],axis=1)

df=df.rename(columns={'higher education':'f_higher education','none':'f_none','secondary education':'f_secondary education'})

df=df.drop('father\'s education',axis=1)
```

```python
df['mother\'s job'].value_counts()
```

```python
df5=pd.get_dummies(df['mother\'s job'],drop_first=True)

df=pd.concat([df5,df],axis=1)

df=df.rename(columns={'civil services':'m_civil services','health care related':'m_health care related','other':'m_other','teacher':'m_teacher'})

df=df.drop('mother\'s job',axis=1)
```

```python
#3. using label encoding for categorical features

df['travel time'].value_counts()
```

```python
le=LabelEncoder()
label=le.fit_transform(df['travel time'])
le.classes_
```

```python
label
```

```python
df=df.drop('travel time',axis=1)
```

```python
df['travel_time']=label
```

```python
df['study time(in hours)'].value_counts()
```

```python
df['extra_curriculum activities '].value_counts()
```

```python
label2=le.fit_transform(df['extra_curriculum activities '])
```

```python
label2
```

```python
df=df.drop('extra_curriculum activities ',axis=1)
```

```python
df['extra_curriculum_activities']=label2
```

```python
df['time spent on internet/social media'].value_counts()
```

```python
label3=le.fit_transform(df['time spent on internet/social media'])
```

```python
label3
```

```python
df=df.drop('time spent on internet/social media',axis=1)
```

```python
df['time_spent_on_internet/socialmedia']=label3
```

```python
df['extra paid classes for courses'].value_counts()
```

```python
df['extra paid classes for courses']=df['extra paid classes for courses'].replace({np.nan:'no'})
```

```python
label4=le.fit_transform(df['extra paid classes for courses'])
```

```python
label4
```

```python
df=df.drop('extra paid classes for courses',axis=1)
```

```python
df['extra_paid_classes_for_courses']=label4
```
